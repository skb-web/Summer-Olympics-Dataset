```python

urlfile="https://raw.githubusercontent.com/AshishJangra27/Skill-India-AI-ML-Scholarship/main/Summer%20Olympics/summer.csv"
df=pd.read_csv(urlfile)
```

# 1. In how many cities Summer Olympics is held so far?

```python
data = []

for city in df['City'].unique():
    data.append(city)
    
print('\n',data,'\n')
print("No. of Cities( =countries) in which Summer Olympics were held is",len(data))
```

# 2. Which sport is having most number of Gold Medals so far?

```python
d=[]
pos=-1
for medal in df['Medal']:
    pos+=1
    if medal =='Gold':
        d.append(pos)
        
s=[]
x=df['Sport']
for t in d:
    s.append(x[t])

r=[ [i,s.count(i)] for i in set(s) ]

def Sort(sub): 
    return(sorted(sub, key = lambda x: x[1]))    

y=Sort(r)
y.reverse()
   
df2 = pd.DataFrame(y, columns =['Sport', 'number']) 
#print(df2)

df2.head().groupby('Sport').mean()['number'].plot(x = 'Sport', y = 'No.of Gold medals',kind = 'bar',color='purple'
                                                  ,title='Top 5 sports having won max no. of gold medals',figsize = (15,5))
```

```python
df2.head(3)
```

### Aquatics (1421) is having most no. of gold medals

# 3. Which sport is having most number of medals so far?

# 6. In which year India won first Gold Medal in Summer Olympics?

```python
d5=[]
pos5=-1
for medal in df['Medal']:
    pos5+=1
    if medal =='Gold':
        d5.append(pos5)
        
x=df['Country']
for t5 in d5:
    if x[t5] =='IND':
        s5=t5
        break

df.iloc[s5]
```

### The year in which India won its first gold medal was 1928 

# 7. Which event is most popular in terms on number of players?

```python
req_data = []
l=0

for event in df['Event'].unique():
    req_data.append([event , len(df[df['Event']  == event])])
    
for i in req_data:
    if i[0]=='Football':
        l+=1        
print(l)
pd.DataFrame(req_data,columns = ['event','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'event', y = 'freq',kind = 'bar',color='orange',title='Top 5 most popular events', figsize = (10,5))
```

### Most popular event is football
